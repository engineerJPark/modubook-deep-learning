```python
#  .

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# X  Y  .
data = [[2, 81], [4, 93], [6, 91], [8, 97]]
x = [i[0] for i in data]
y = [i[1] for i in data]

#  .
plt.figure(figsize=(8,5))
plt.scatter(x, y)
plt.show()

```

```python
#   x y    .(        .)
x_data = np.array(x)
y_data = np.array(y)

#  a  b   .
a = 0
b = 0

# .
lr = 0.03 

#   .
epochs = 2001 

#  .
for i in range(epochs): # epoch   
    y_hat = a * x_data + b  #y   
    error = y_data - y_hat  #  .
    a_diff = -(2/len(x_data)) * sum(x_data * (error)) #  a  . 
    b_diff = -(2/len(x_data)) * sum(error)  #  b  . 
    a = a - lr * a_diff  #    a .
    b = b - lr * b_diff  #    b .
    if i % 100 == 0:    # 100    a, b .
        print("epoch=%.f, =%.04f, =%.04f" % (i, a, b))
```

```python
#        .
y_pred = a * x_data + b
plt.scatter(x, y)
plt.plot([min(x_data), max(x_data)], [min(y_pred), max(y_pred)])
plt.show()
```
